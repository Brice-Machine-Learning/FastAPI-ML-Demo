## Interpreting the Model Input

When calling the `/predict` endpoint, the request payload looked like this:

```json
{
  "features": [6.0, 25.0, 6.0, 1.0, 900, 3.0, 34.05, -118.25]
}
The `features` array contains 8 values, each corresponding to a specific feature used by the model. The order of these features is crucial, as the model expects them in a predefined sequence.
```
The order of features is as follows:
1. `MedInc`: Median income in block group *(in tens of thousands of dollars)*
2. `HouseAge`: Median house age in block group *(years)*
3. `AveRooms`: Average number of rooms per household
4. `AveBedrms`: Average number of bedrooms per household
5. `Population`: Block group population *(census tract: number of people.  Could be within a block group or neighborhood)*
6. `AveOccup`: Average number of household members
7. `Latitude`: Block group latitude *(degrees)*
8. `Longitude`: Block group longitude *(degrees)*
Each value in the `features` array corresponds to the respective feature in the order listed above. For example, in the provided payload:
- `6.0` corresponds to `MedInc`
- `25.0` corresponds to `HouseAge`
- `6.0` corresponds to `AveRooms`
- `1.0` corresponds to `AveBedrms`
- `900` corresponds to `Population`
- `3.0` corresponds to `AveOccup`
- `34.05` corresponds to `Latitude`
- `-118.25` corresponds to `Longitude`
When preparing the input for the model, ensure that the features are provided in this exact order to guarantee accurate predictions.
